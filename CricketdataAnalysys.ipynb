```python
odi_bowlers=pd.read_csv('Bowling/Bowling_ODI.csv')
odi_bowlers.head()
```

```python
odi_bowlers.drop(["Unnamed: 0","Unnamed: 13"],axis=1,inplace=True)
```

```python
odi_bowlers.info()
```

```python
import numpy as np

odi_bowlers["Inns"] = odi_bowlers["Inns"].replace("-", np.nan).astype(float).astype("Int64")
odi_bowlers["Wkts"] = odi_bowlers["Wkts"].replace("-", np.nan).astype(float).astype("Int64")
odi_bowlers["Runs"] = odi_bowlers["Runs"].replace("-", np.nan).astype(float).astype("Int64")

# Keep BBI as string (or split later if needed)
odi_bowlers["BBI"] = odi_bowlers["BBI"].astype(str).str.replace("*", "", regex=False).replace("-", np.nan)

odi_bowlers["Ave"] = odi_bowlers["Ave"].replace("-", np.nan).astype(float)
odi_bowlers["4"] = odi_bowlers["4"].replace("-", np.nan).astype(float).astype("Int64")
odi_bowlers["5"] = odi_bowlers["5"].replace("-", np.nan).astype(float).astype("Int64")
odi_bowlers["Econ"] = odi_bowlers["Econ"].replace("-", np.nan).astype(float)  # Don't cast to Int64
```

```python
odi_bowlers[['Stariting_year', 'Ending_year']] = odi_bowlers['Span'].str.split('-', expand=True)
odi_bowlers["Stariting_year"] = odi_bowlers["Stariting_year"].replace("-", np.nan).astype(float).astype("Int64")
odi_bowlers["Ending_year"] = odi_bowlers["Ending_year"].replace("-", np.nan).astype(float).astype("Int64")
```

```python
odi_bowlers.drop("Span",axis=1,inplace=True)
```

```python
# Ensure all values in 'Player' column are strings
odi_bowlers['Player'] = odi_bowlers['Player'].astype(str)

# Split 'Player' column on first '(' only
split1_columns = odi_bowlers['Player'].str.split(r'\(', n=1, expand=True)

# Check if split1_columns[1] exists before modification
split1_columns[1] = split1_columns[1].fillna("").str.replace(')', '', regex=False).str.strip()

# Assign properly split values
odi_bowlers["player"] = split1_columns[0].str.strip()
odi_bowlers["country"] = split1_columns[1].str.replace(r'Asia|ICC|/', '', regex=True).str.strip()

# Drop old 'Player' column
odi_bowlers.drop(columns=['Player'], inplace=True)

# Check mismatches
print(odi_bowlers[['player', 'country']].head(10))
```

```python
odi_bowlers.head()
```

```python
odi_bowlers["country"].unique()
```

```python
# Fix country names by mapping weird combos to the right name
country_map = {
    "SL": "Sri Lanka",
    "PAK": "Pakistan",
    "INDIA": "India",
    "AUS": "Australia",
    "NZ": "New Zealand",
    "SA": "South Africa",
    "ENG": "England",
    "BDESH": "Bangladesh",
    "WI": "West Indies",
    "ZIM": "Zimbabwe",
    "AFG": "Afghanistan",
    "IRE": "Ireland",
    "KENYA": "Kenya",
    "SCOT": "Scotland",
    "NL": "Netherlands",
    "UAE": "UAE",
    "CAN": "Canada",
    "HKG": "Hong Kong",
    "BMUDA": "Bermuda",
    "PNG": "Papua New Guinea",
    "NAM": "Namibia",
    "USA": "USA",
    "NEPAL": "Nepal",
    "OMAN": "Oman",
    
    # Fixes for hybrid/messy entries
    "AfrSA": "South Africa",
    "AfrZIM": "Zimbabwe",
    "AfrKENYA": "Kenya",
    "ENGIRE": "England",
    "CANWI": "West Indies",
    "ENGSCOT": "England",
    "NLSA": "South Africa",
    "AUSSA": "Australia",
    "3 (PAK": "Pakistan",
    "SAUSA": "South Africa",
    "USAWI": "West Indies",
    "EAf": "South Africa",
    "1 (UAE": "UAE",
    "HKGNZ": "Hong Kong",
    "ENGPNG": "England",
    "AUSNZ": "Australia"
}

# Apply mapping
odi_bowlers['country'] = odi_bowlers['country'].map(country_map).fillna(odi_bowlers['country'])
odi_bowlers.head()
```
